```
from imageai.Detection.Custom import DetectionModelTrainer

trainer = DetectionModelTrainer()
trainer.setModelTypeAsYOLOv3()
trainer.setDataDirectory(data_directory="/content/drive/My Drive/Data_Investment_Dataset/DATA")
trainer.setTrainConfig(object_names_array=["AMERICAN FUNDS", "FIDELITY", "EDWARD JONES", "MERRILL LYNCH", ], batch_size=16, num_experiments=50, train_from_pretrained_model="/content/pretrained-yolov3.h5")
trainer.trainModel()
```

```
from imageai.Detection.Custom import DetectionModelTrainer

trainer = DetectionModelTrainer()
trainer.setModelTypeAsYOLOv3()
trainer.setDataDirectory(data_directory="/content/drive/My Drive/Data_Investment_Dataset/DATA")
trainer.evaluateModel(model_path="/content/drive/My Drive/Data_Investment_Dataset/DATA/models", json_path="/content/drive/My Drive/Data_Investment_Dataset/DATA/json/detection_config.json", iou_threshold=0.5, object_threshold=0.3, nms_threshold=0.5)
```

```
from imageai.Detection.Custom import CustomObjectDetection

detector = CustomObjectDetection()
detector.setModelTypeAsYOLOv3()
name = []
detector.setModelPath("/content/drive/My Drive/Data_Investment_Dataset/DATA/models/detection_model-ex-009--loss-0011.470.h5") 
detector.setJsonPath("/content/drive/My Drive/Data_Investment_Dataset/DATA/json/detection_config.json")
detector.loadModel()
detections = detector.detectObjectsFromImage(input_image="/5 - Copy (2).jpg", output_image_path="logo_detected_image.jpg")
for detection in detections:
    print(detection["name"], " : ", detection["percentage_probability"], " : ", detection["box_points"])
    name.append(detection["name"])
```

```
from PIL import Image

image = Image.open('logo_detected_image.jpg')
image = image.resize((500, 500))
display(image)
print()
print()
print("The application form given, belongs to", name[0])
```
